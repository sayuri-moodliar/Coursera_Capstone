## Scrape the Wikipedia page for postal codes of Toronto, Canada

```python
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
bs=BeautifulSoup(source,'lxml')
print(bs.title)
tab = str(bs.table)
display_html(tab,raw=True)
```

## Create a pandas dataframe

```python
#Convert the table to a dataframe for preprocessing and cleaning
df_bs = pd.read_html(tab)
df=df_bs[0]
df.head()
```

```python
#Drop rows containing cells with Borough 'Not assigned'
df1 = df[df.Borough != 'Not assigned']

#Combine rows containing neighborhoods with same postal code
df2 = df1.groupby(['Postal Code','Borough'], sort=False).agg(','.join)
df2.reset_index(inplace=True)

#Change names of 'Not assigned' Neighborhood to Borough
df2['Neighbourhood'] = np.where(df2['Neighbourhood'] == 'Not assigned', df2['Borough'], df2['Neighbourhood'])

df2
```

```python
#Print number of rows and columns of dataframe
df2.shape
```

## EXERCISE 2

## Create dataframe of latitudes and longitudes of neighborhoods in Toronto

```python
#Import csv file with coordinates
lat_lng = pd.read_csv('http://cocl.us/Geospatial_data')
lat_lng.head()
```

```python
#Merge Neighborhoods df with Coordinates df
df3 = pd.merge(df2,lat_lng,on='Postal Code')
df3.head(12)
```

## EXERCISE 3

## Generate a map of clustered neighborhoods in Toronto

```python
#Get rows from df which contain Toronto in Borough
df4=df3[df3['Borough'].str.contains('Toronto')]
df4
```

```python
#Create Toronto map to visualize neighborhoods
map_toronto = folium.Map(location=[43.651070,-79347015],zoom_start=10)

for lat,lng,borough,neighborhood in zip(df4['Latitude'],df4['Longitude'],df4['Borough'],df4['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat,lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)

map_toronto
```

```python
#Use KMeans clustering to cluster neighborhoods
k=5
toronto_clustering = df4.drop(['Postal Code','Borough','Neighbourhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
df4.insert(0, 'Cluster Labels', kmeans.labels_)
df4
```

```python
#Create map with clusters
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]
markers_colors = []
for lat, lng, neighborhood, cluster in zip(df4['Latitude'], df4['Longitude'], df4['Neighbourhood'], df4['Cluster Labels']):
    label = folium.Popup('Cluster' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat,lng],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
    
map_clusters
```
